```python
# Convertir la columna de ventas totales (ingresos_totales) a un array de Numpy
ventas_array = df_ventas['ingresos_totales'].to_numpy()

# Calcular la mediana y la desviación estándar utilizando Numpy
mediana_ventas = np.median(ventas_array)
std_ventas = np.std(ventas_array)

print(f"Ventas Totales (Numpy Array): {ventas_array}")
print(f"Mediana de ventas (Numpy): {mediana_ventas}")
print(f"Desviación estándar de ventas (Numpy): {std_ventas}")
```

```python
# Simulación de proyecciones de ventas futuras mediante arrays aleatorios de Numpy
np.random.seed(42)  # Establecer semilla para garantizar reproducibilidad

# Generar proyecciones basadas en la media y desviación estándar obtenidas anteriormente
proyecciones = np.random.normal(loc=mediana_ventas, scale=std_ventas, size=1000)

# Calcular estadísticas básicas sobre las proyecciones
media_proyectada = np.mean(proyecciones)
mediana_proyectada = np.median(proyecciones)
percentiles_proyectados = np.percentile(proyecciones, [25, 50, 75])

print(f"--- Proyecciones de Ventas Futuras ---")
print(f"Media proyectada: {media_proyectada}")
print(f"Mediana proyectada: {mediana_proyectada}")
print(f"Percentiles (25, 50, 75): {percentiles_proyectados}")
```

4. Análisis de inventarios (Pandas)

- Calcula la rotación de inventarios para cada tienda. Esto se hace dividiendo las ventas totales por el stock disponible de cada producto.

- Almacena los resultados en una nueva columna dentro del DataFrame de inventarios.

- Filtra y muestra las tiendas con niveles críticos de inventario, es decir, aquellas tiendas donde el porcentaje de productos vendidos sea menor al 10% del stock disponible.

- Utiliza groupby() y operaciones matemáticas para calcular la rotación de inventarios.

- Aplica filtros con Pandas para identificar las tiendas con niveles críticos.

```python
# rotacion de inventario: alinear por ID_Tienda y Producto
ventas_agregadas = df_ventas.groupby(['ID_Tienda', 'Producto'])['Cantidad_Vendida'].sum().reset_index()
df_inventarios = df_inventarios.merge(ventas_agregadas, on=['ID_Tienda', 'Producto'], how='left')
df_inventarios['rotacion_inventario'] = df_inventarios['Cantidad_Vendida'] / df_inventarios['Stock_Disponible']
df_inventarios
```

```python
# Calcular la rotación de inventarios por tienda de forma coherente (promedio por producto)
df = df_inventarios.groupby('ID_Tienda')['rotacion_inventario'].mean()
df = df.to_frame().copy()
df
```

```python
# tiendas con rotacion de inventario menor a 10%
df['Estado'] = np.where(df['rotacion_inventario'] <= 0.10, 'Baja', 'Alta')
for row in df.itertuples():
    if row.Estado == 'Baja':
        print ('La tienda', row.Index, 'tiene una rotacion de inventario menor de 10%') 
    else:
        print ('La tienda',row.Index, 'tiene una rotacion de inventario mayor de 10%')
```
